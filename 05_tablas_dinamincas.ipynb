# Agregaciones Avanzadas y Tablas Dinámicas

```python
import pandas as pd
import numpy as np
```

```python
#Inicializado de un dataset de 2000 registros de rendimientos en tienda.

np.random.seed(42)
n = 2000

df_tiendas = pd.DataFrame({
    'tienda': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], n),
    'gerente': np.random.choice(['Ana', 'Carlos', 'Elena', 'Luis'], n),
    'producto': np.random.choice(['Dispositivos', 'Accesorios', 'Servicios'], n),
    'ventas': np.random.uniform(100, 5000, n).round(2),
    'satisfaccion': np.random.randint(1, 6, n)
})
print(f"Dataset de rendimiento listo con {len(df_tiendas)} registros.")
```

## Generación de Tablas dinámicas ( .pivot_table( index = , columns = , values = ) )
index = define las filas de la tabla dinámica.

columns = define las variables/columnas de la tabla dinámica.

values = define los valores numéricos que se quieren usar para el cálculo.

aggfunc = le dice a la función las operaciones que tiene que llevar a cabo sobre los datos.(por defecto haría la media)

```python
#Creo una tabla dinámica con pivot_table()
tabla_ventas = df_tiendas.pivot_table(index = 'tienda', columns = 'producto', values = 'ventas', aggfunc = 'sum')
print(tabla_ventas)
```

```python
#Tabal con la suma de las marginales
tabla_con_totales = df_tiendas.pivot_table(
    index = 'tienda',
    columns = 'producto',
    values = 'ventas',
    aggfunc = 'sum',
    margins = True,
    margins_name= 'Total'
)
print(tabla_con_totales)
```

## Tablas dinámicas con índices múltiples


```python
tabla_multi = df_tiendas.pivot_table(
    index = ['tienda', 'gerente'],
    columns = 'producto',
    values = 'ventas',
    aggfunc = 'mean',
)
print(tabla_multi)
```
