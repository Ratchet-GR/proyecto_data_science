# Visualización de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #biblioteca con las funciones plot para representación matemática
```

```python
#Generamos la misma tabla que en el notebook anterior
np.random.seed(42)
n = 2000
df_tiendas = pd.DataFrame({
    'tienda': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], n),
    'producto': np.random.choice(['Dispositivos', 'Accesorios', 'Servicios'], n),
    'ventas': np.random.uniform(100, 5000, n).round(2),
    'fecha': pd.date_range(start='2026-01-01', periods=n, freq='h')
})
tabla_grafico = df_tiendas.pivot_table(index='tienda', columns='producto', values='ventas', aggfunc='sum')
print(tabla_grafico)
```

## Generar Diagramas de Barras con Matplotlib

```python
#Generación de un Gráfico de barras:
tabla_grafico.plot(kind = 'bar', figsize = (10,6))
plt.title('Ventas totales por tienda y producto')
plt.ylabel('Ventas')
plt.xlabel('tiendas')
plt.show
```

### Gráfico de Barras apiladas

```python
tabla_grafico.plot(kind = 'bar', stacked = True, figsize = (10,6))
plt.title('Ventas totales por tienda y producto')
plt.ylabel('Ventas')
plt.xlabel('tiendas')
plt.show
```

```python
# Agrupamos las ventas por fecha para ver la tendencia temporal
tendencia_ventas = df_tiendas.groupby('fecha')['ventas'].sum()

# Graficamos la tendencia
tendencia_ventas.plot(kind='line', figsize=(12, 5), color='teal')
plt.title('Tendencia de Ventas en el Tiempo')
plt.ylabel('Total Ventas')
plt.xlabel('Fecha/Hora')
plt.grid(True) # Añade una cuadrícula para facilitar la lectura
plt.show()
```
