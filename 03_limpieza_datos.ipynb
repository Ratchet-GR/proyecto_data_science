# Limpieza de Datos

```python
import pandas as pd
import numpy as np
```

```python
#Inicializado de un Dataframe con información faltante
np.random.seed(24)
n = 1200

df_sucio = pd.DataFrame({
    'id_usuario': range(1001, 1001 + n),
    'edad': np.random.choice([22, 35, 45, np.nan, 28, 55, np.nan], n),
    'ingresos': np.random.choice([25000, 40000, 55000, np.nan, 85000, np.nan], n),
    'ciudad': np.random.choice(['Madrid', 'Barcelona', 'Valencia', None, 'Sevilla'], n)
})

# Duplicamos las primeras 50 filas para simular un error de carga de datos
df_sucio = pd.concat([df_sucio, df_sucio.iloc[:50]], ignore_index=True)
print(f"Dataset creado con {len(df_sucio)} registros.")
```

```python
#Ver qué cantidad de valores son nulos en cada una de las variables
df_sucio.isnull().sum()
```

```python
#Ver qué cantidad de filas duplicadas hay en el dataframe.
df_sucio.duplicated().sum()
```

```python
df_sucio.drop_duplicates(inplace = True)
```

```python
len(df_sucio)
```

### Eliminar filas con datos NaN en variables categóricas

```python
df_sucio.dropna(subset = ['ciudad'], inplace = True)
```

```python
len(df_sucio)
```

### Imputación de datos numéricos con la mediana o la media

```python
#Cálculo de la mediana de la edad
mediana_edad = df_sucio['edad'].median()
#Imputación de la edad medianan
df_sucio['edad'].fillna(mediana_edad,  inplace = True)

#Cálculo de la media de ingresos
media_ingresos = df_sucio['ingresos'].mean()
#Imputamos los NaN de ingresos con la media
df_sucio['ingresos'].fillna(media_ingresos, inplace = True)
```

```python
df_sucio.isnull().sum()
```

```python
print(df_sucio.dtypes)
```

```python
df_sucio['edad'] = df_sucio['edad'].astype(int)
df_sucio['ingresos'] = df_sucio['ingresos'].astype(int)
```

### Consistencia de Texto
Hay que escoger un formato en el que presentar el texto de una variable categórica, si no, se podrían confundir 
debido al uso de mayúsculas-minúsculas o espacios

```python
df_sucio['ciudad'] = df_sucio['ciudad'].str.strip().str.title()
```

### Detección de Valores Atípicos

```python
df_sucio[['edad', 'ingresos']].describe()
```

```python
#Ingresos
Cuart_1 = df_sucio['ingresos'].quantile(0.25)
Cuart_3 = df_sucio['ingresos'].quantile(0.75)
iqr = Cuart_3 - Cuart_1

#Límites para outliers
lim_sup = Cuart_3 + 1.5 * iqr
lim_inf = Cuart_1 - 1.5 * iqr

df_final = df_sucio[(df_sucio['ingresos'] <= lim_sup) & (df_sucio['ingresos'] >= lim_inf)]

print(f"Los registros antes de eliminar outliers son {len(df_sucio)}")
print(f"Los registros después de eliminar outliers son {len(df_final)}")
```
