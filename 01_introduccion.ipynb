## Numpy

La operación que hemos realizado no ha multiplicado los elementos de la lista "marks" por 2, sino que,
debido a la forma que tiene python de ver la listas, como un objeto en memoria y no algebraico, lo que ha hecho es duplicar la lista.

Para atajar este problema se creó la biblioteca Numpy, para poder trabajar con álgebra lineal a través de su estructura de "array"
con la que podremos definir vectores y matrices y trabajar con ellos de forma cómoda.

```python
# Importamos la biblioteca Numpy
import numpy as np

# Convertimos la lista "marks" en un array de numpy
marks_vector = np.array(marks)

# Operamos con una sitaxis similar a la que hemos intentado utilizar previamente
double_marks_vector = marks_vector * 2

# Imprimimos en pantalla el resultado de la operción
print("Las notas iniciales eran: ", marks_vector)
print("Las notas duplicadas son: ", double_marks_vector)
```

Una vez visto cómo trabajan, brevemente, los arrays unidimensionales (vectores) de numpy vamos a ver cómo trabajar con arrays bidimensionales, o matrices. 
Este tipo de estructuras de datos se corresponden con los datasets comunes en los que cada fila es una medición de algún elemento de una población (o registro),
y cada columna representa los distintos registros de una variable medible, sea del tipo que sea cualitativa o cuantitativa (booleana, de enteros, decimal pto. flotante, etc).

```python
# Definición de una matriz como lista de listas
A_matrix = np.array([
    [10, 20, 30], # Fila 0 (row 1)
    [40, 50, 60]  # Fila 1 (row 2)
])

print("La matriz definida es: ", A_matrix)
print("\nLa dimensión de la matriz es: ", A_matrix.shape) # array.shape devuelve una dupla con las dimensiones de la matriz
```

```python
#Imprimimos un dato específico de la matriz "A_matrix"
print(A_matrix[1, 1])
```

```python
#Vamos a sustraer una parte de una matriz, en este caso, la segunda columna
column_2 = A_matrix[:, 1]

print("La segunda columna es: ", column_2)
```

```python
#Vamos a sustraer una parte de una matriz, en este caso, la segunda columna
column_1 = A_matrix[:, 0]

print("La segunda columna es: ", column_1)
```

En la siguiente caja de código vamos a crear una máscara booleana, esto es una condición, verdadera o 
falsa y se la vamos a aplicar a la matriz para filtrar datos en base a la condición de la máscara.

```python
#Matriz:
# [10, 20, 30]
# [40, 50, 60]

#Creamos la condición y la almacenamos en una máscara booleana
mask = A_matrix > 25

print("1. Matriz de booleanos: \n", mask)

#Filtrramos los datos de la matriz en base a la matriz booleana que hemos generado
filtered_A_matrix = A_matrix[mask]

print("\nLos elementos que cumplen la condición son: \n", filtered_A_matrix)
#Pérdida de dimensionalidad por filtrado
```

Una vez llegados a este punto tenemos dos opciones:

### Deteción matemática de outliers para una variable
El IQR es el rango intercuartílico (Q1-Q3)

La Regla de los bigotes de Turkey nos dice que un dato es outlier si:
- Es menor que Q1 - 1.5*IQR
- ES mayor que que Q3 + 1.5*IQR

```python
#Cálculo de los cuartiles mediante la función .quantile( entre [0,1] )
q1 = df_pacientes['edad'].quantile(0.25)
q3 = df_pacientes['edad'].quantile(0.75)

iqr = q3 - q1

limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

df_pacientes[(df_pacientes['edad'] < limite_inferior) | (df_pacientes['edad'] > limite_superior)]
```

#### Una vez identificados los outliers tenemos 3 posibilidades:
- Corregir: Si descubrimos que fue un error de dedo (como escribir 220 en lugar de 22), 
se arregla con el valor real.
- Acotar (Winsorización): Se reemplaza el valor extremo por el valor del limite_superior 
o limite_inferior para que no distorsione los extremos, pero sin perder la fila.
- Eliminar: Solo si se demuestra que es un fallo del equipo de medición o ruido incorregible.


### Uso de la imputación mediante el algoritmo KNN
Este algoritmo toma los n valores más próximos en una variable e imputa la media de otra 
variable al valor NaN.

```python
#Volvemos a asignar un valor NaN a una de las celdas del DataFrame
df_pacientes.loc['Marta','presion_arterial'] = np.nan
df_pacientes
```

```python
#Añadimos el archivo impute de la librería skit-learn para imputar mediante KNN
from sklearn.impute import KNNImputer

#Configuramos el imputador
imputer = KNNImputer(n_neighbors = 2)

#El algoritmo entrena y rellena los valores automáticamente
df_valores_limpios = imputer.fit_transform(df_pacientes)
```

```python
df_valores_limpios
```

```python
#Reconstruimos nuestro dataframe de Pandas puesto que ha arrojado un array de Numpy
df_valores_limpios = pd.DataFrame(
    df_valores_limpios,
    columns = df_pacientes.columns,
    index = df_pacientes.index
)
```
